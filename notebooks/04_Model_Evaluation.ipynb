```python
# ==========================================================
# NOTEBOOK 04
# MODEL EVALUATION
# ==========================================================

"""
OBJECTIVE
---------
Evaluate the trained Custom CNN model using the
test dataset.

This notebook performs:

1. Load trained model
2. Load test dataset
3. Predict classes
4. Calculate accuracy
5. Generate confusion matrix
6. Generate classification report
7. Plot ROC curves
8. Save evaluation graphs
"""

# ==========================================================
# IMPORT LIBRARIES
# ==========================================================

import os
import shutil
import zipfile
import pickle

import numpy as np
import pandas as pd
import tensorflow as tf

import matplotlib.pyplot as plt

from google.colab import drive

from tensorflow.keras.models import load_model

from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    accuracy_score,
    ConfusionMatrixDisplay,
    roc_curve,
    auc
)

from sklearn.preprocessing import label_binarize

print("="*60)
print("TensorFlow Version :", tf.__version__)
print("GPU :", tf.config.list_physical_devices("GPU"))
print("="*60)

# ==========================================================
# PROJECT SETTINGS
# ==========================================================

IMAGE_SIZE = (224,224)

BATCH_SIZE = 32

AUTOTUNE = tf.data.AUTOTUNE

# ==========================================================
# MOUNT GOOGLE DRIVE
# ==========================================================

drive.mount('/content/drive')

# ==========================================================
# PROJECT PATHS
# ==========================================================

PROJECT_PATH="/content/drive/MyDrive/Covid_Major_Project"

PROCESSED_PATH=os.path.join(
    PROJECT_PATH,
    "processed"
)

MODEL_PATH=os.path.join(
    PROJECT_PATH,
    "models"
)

GRAPH_PATH=os.path.join(
    PROJECT_PATH,
    "graphs"
)

os.makedirs(GRAPH_PATH,exist_ok=True)

# ==========================================================
# LOAD TRAINED MODEL
# ==========================================================

model=load_model(

    os.path.join(
        MODEL_PATH,
        "best_custom_cnn.keras"
    )

)

print()

print("="*60)
print("MODEL LOADED SUCCESSFULLY")
print("="*60)

model.summary()

# ==========================================================
# EXTRACT DATASET
# ==========================================================

ZIP_FILE = os.path.join(
    PROJECT_PATH,
    "dataset",
    "archive.zip"
)

TEMP_FOLDER = "/content/COVID19_Dataset"

if os.path.exists(TEMP_FOLDER):
    shutil.rmtree(TEMP_FOLDER)

os.makedirs(TEMP_FOLDER, exist_ok=True)

print("\nExtracting Dataset...")

with zipfile.ZipFile(ZIP_FILE, "r") as zip_ref:
    zip_ref.extractall(TEMP_FOLDER)

print("Dataset Extracted Successfully.")

DATASET_PATH = os.path.join(
    TEMP_FOLDER,
    "COVID-19_Radiography_Dataset"
)

# ==========================================================
# LOAD TEST DATA
# ==========================================================

test_df = pd.read_csv(
    os.path.join(PROCESSED_PATH, "test.csv")
)

def fix_path(row):

    filename = os.path.basename(row["Image"])

    return os.path.join(
        DATASET_PATH,
        row["Label"],
        "images",
        filename
    )

test_df["Image"] = test_df.apply(
    fix_path,
    axis=1
)

with open(
    os.path.join(PROCESSED_PATH, "label_encoder.pkl"),
    "rb"
) as file:
    label_encoder = pickle.load(file)

print("="*60)
print("VERIFY IMAGE PATHS")
print("="*60)

print(test_df[["Label", "Image"]].head())

print()

missing = (~test_df["Image"].apply(os.path.exists)).sum()

print("Missing files :", missing)

print("="*60)

print()

print("="*60)
print("TEST DATASET")
print("="*60)

print(test_df.head())

print()

print("Total Test Images :", len(test_df))

print()

print("Classes")

for cls in label_encoder.classes_:
    print(cls)

# ==========================================================
# IMAGE PREPROCESSING FUNCTION
# ==========================================================

def preprocess_image(image_path, label):

    image = tf.io.read_file(image_path)

    image = tf.image.decode_png(
        image,
        channels=3
    )

    image = tf.image.resize(
        image,
        IMAGE_SIZE
    )

    image = tf.cast(
        image,
        tf.float32
    ) / 255.0

    return image, label

# ==========================================================
# CREATE TEST DATASET
# ==========================================================

test_dataset = tf.data.Dataset.from_tensor_slices(

    (
        test_df["Image"].values,
        test_df["Label_ID"].values
    )

)

test_dataset = test_dataset.map(

    preprocess_image,

    num_parallel_calls=AUTOTUNE

)

test_dataset = test_dataset.batch(BATCH_SIZE)

test_dataset = test_dataset.prefetch(AUTOTUNE)

print()

print("="*60)
print("TEST DATASET CREATED SUCCESSFULLY")
print("="*60)

# ==========================================================
# MAKE PREDICTIONS
# ==========================================================

"""
This section loads all test images into the trained model
and predicts their classes.

The predicted labels are then compared with the true labels
to calculate the overall test accuracy.
"""

print("="*60)
print("PREDICTING TEST IMAGES")
print("="*60)

# Predict probabilities
predictions = model.predict(
    test_dataset,
    verbose=1
)

# Convert probabilities to class IDs
predicted_labels = np.argmax(
    predictions,
    axis=1
)

# Actual labels
true_labels = test_df["Label_ID"].values

print()

print("Prediction Completed Successfully.")

print()

print("Total Predictions :", len(predicted_labels))

# ==========================================================
# CALCULATE TEST ACCURACY
# ==========================================================

test_accuracy = accuracy_score(

    true_labels,

    predicted_labels

)

print("="*60)
print("TEST ACCURACY")
print("="*60)

print(f"Test Accuracy : {test_accuracy:.4f}")

print(f"Test Accuracy : {test_accuracy*100:.2f}%")

# ==========================================================
# DISPLAY SAMPLE PREDICTIONS
# ==========================================================

print("="*60)
print("SAMPLE PREDICTIONS")
print("="*60)

for i in range(10):

    actual = label_encoder.inverse_transform(
        [true_labels[i]]
    )[0]

    predicted = label_encoder.inverse_transform(
        [predicted_labels[i]]
    )[0]

    print(f"Image {i+1}")

    print("Actual    :", actual)

    print("Predicted :", predicted)

    print("-"*40)

    # ==========================================================
# PREDICTION SUMMARY
# ==========================================================

print("="*60)
print("PREDICTION SUMMARY")
print("="*60)

correct = np.sum(
    true_labels == predicted_labels
)

incorrect = len(true_labels) - correct

print("Correct Predictions   :", correct)

print("Incorrect Predictions :", incorrect)

print("Total Images          :", len(true_labels))

print()

print(f"Overall Accuracy      : {test_accuracy*100:.2f}%")

# ==========================================================
# CONFUSION MATRIX
# ==========================================================

"""
A confusion matrix shows how many images were
correctly and incorrectly classified for each class.
"""

cm = confusion_matrix(

    true_labels,

    predicted_labels

)

print("="*60)
print("CONFUSION MATRIX")
print("="*60)

print(cm)

# ==========================================================
# DISPLAY CONFUSION MATRIX
# ==========================================================

plt.figure(figsize=(8,8))

disp = ConfusionMatrixDisplay(

    confusion_matrix=cm,

    display_labels=label_encoder.classes_

)

disp.plot(

    cmap="Blues",

    values_format="d"

)

plt.title("Confusion Matrix - Custom CNN")

plt.xticks(rotation=20)

plt.tight_layout()

plt.show()

# ==========================================================
# SAVE CONFUSION MATRIX
# ==========================================================

plt.figure(figsize=(8,8))

disp = ConfusionMatrixDisplay(

    confusion_matrix=cm,

    display_labels=label_encoder.classes_

)

disp.plot(

    cmap="Blues",

    values_format="d"

)

plt.title("Confusion Matrix - Custom CNN")

plt.xticks(rotation=20)

plt.tight_layout()

plt.savefig(

    os.path.join(

        GRAPH_PATH,

        "custom_confusion_matrix.png"

    ),

    dpi=300,

    bbox_inches="tight"

)

plt.close()

print()

print("Confusion Matrix saved successfully.")

# ==========================================================
# CLASS-WISE ACCURACY
# ==========================================================

print("="*60)
print("CLASS-WISE CORRECT PREDICTIONS")
print("="*60)

for i, class_name in enumerate(label_encoder.classes_):

    total = np.sum(cm[i])

    correct = cm[i][i]

    accuracy = (correct / total) * 100

    print(f"{class_name:20s} : {accuracy:.2f}%")

# ==========================================================
# SUMMARY
# ==========================================================

print()

print("="*60)
print("CONFUSION MATRIX COMPLETED")
print("="*60)

print("Graph saved as:")

print(

    os.path.join(

        GRAPH_PATH,

        "custom_confusion_matrix.png"

    )

)

# ==========================================================
# CLASSIFICATION REPORT
# ==========================================================

"""
The classification report provides:

1. Precision
2. Recall
3. F1-Score
4. Support

for each disease class.
"""

print("="*60)
print("CLASSIFICATION REPORT")
print("="*60)

report = classification_report(

    true_labels,

    predicted_labels,

    target_names=label_encoder.classes_,

    digits=4

)

print(report)

# ==========================================================
# SAVE CLASSIFICATION REPORT
# ==========================================================

report_file = os.path.join(

    MODEL_PATH,

    "classification_report.txt"

)

with open(report_file, "w") as file:

    file.write(report)

print()

print("Classification Report Saved Successfully.")

print(report_file)

# ==========================================================
# CREATE DATAFRAME OF REPORT
# ==========================================================

report_dict = classification_report(

    true_labels,

    predicted_labels,

    target_names=label_encoder.classes_,

    output_dict=True

)

report_df = pd.DataFrame(report_dict).transpose()

print("="*60)
print("CLASSIFICATION REPORT DATAFRAME")
print("="*60)

print(report_df)

# ==========================================================
# SAVE REPORT AS CSV
# ==========================================================

csv_file = os.path.join(

    MODEL_PATH,

    "classification_report.csv"

)

report_df.to_csv(csv_file)

print()

print("Classification Report CSV Saved Successfully.")

print(csv_file)

# ==========================================================
# DISPLAY OVERALL METRICS
# ==========================================================

print("="*60)
print("OVERALL MODEL PERFORMANCE")
print("="*60)

print(f"Overall Accuracy : {test_accuracy:.4f}")

print()

print("Macro Average")

print("------------------------")

print(f"Precision : {report_dict['macro avg']['precision']:.4f}")

print(f"Recall    : {report_dict['macro avg']['recall']:.4f}")

print(f"F1 Score  : {report_dict['macro avg']['f1-score']:.4f}")

print()

print("Weighted Average")

print("------------------------")

print(f"Precision : {report_dict['weighted avg']['precision']:.4f}")

print(f"Recall    : {report_dict['weighted avg']['recall']:.4f}")

print(f"F1 Score  : {report_dict['weighted avg']['f1-score']:.4f}")

# ==========================================================
# ROC CURVE
# ==========================================================

"""
ROC (Receiver Operating Characteristic) curves are plotted
for each class using the prediction probabilities.
"""

# Convert labels to one-hot encoding
true_labels_bin = label_binarize(
    true_labels,
    classes=[0,1,2,3]
)

plt.figure(figsize=(8,6))

for i in range(len(label_encoder.classes_)):

    fpr, tpr, _ = roc_curve(
        true_labels_bin[:,i],
        predictions[:,i]
    )

    roc_auc = auc(fpr, tpr)

    plt.plot(
        fpr,
        tpr,
        linewidth=2,
        label=f"{label_encoder.classes_[i]} (AUC = {roc_auc:.2f})"
    )

plt.plot(
    [0,1],
    [0,1],
    linestyle="--",
    linewidth=2
)

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("ROC Curves - Custom CNN")

plt.legend(loc="lower right")

plt.grid(True)

plt.tight_layout()

plt.show()

# ==========================================================
# SAVE ROC CURVE
# ==========================================================

plt.figure(figsize=(8,6))

for i in range(len(label_encoder.classes_)):

    fpr, tpr, _ = roc_curve(
        true_labels_bin[:,i],
        predictions[:,i]
    )

    roc_auc = auc(fpr, tpr)

    plt.plot(
        fpr,
        tpr,
        linewidth=2,
        label=f"{label_encoder.classes_[i]} (AUC={roc_auc:.2f})"
    )

plt.plot(
    [0,1],
    [0,1],
    linestyle="--"
)

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("ROC Curves - Custom CNN")

plt.legend(loc="lower right")

plt.grid(True)

plt.tight_layout()

roc_file = os.path.join(
    GRAPH_PATH,
    "custom_roc_curve.png"
)

plt.savefig(
    roc_file,
    dpi=300,
    bbox_inches="tight"
)

plt.close()

print("ROC Curve Saved Successfully.")

# ==========================================================
# VERIFY SAVED FILES
# ==========================================================

print("="*60)
print("VERIFYING OUTPUT FILES")
print("="*60)

files = [

    "custom_accuracy.png",

    "custom_loss.png",

    "custom_confusion_matrix.png",

    "custom_roc_curve.png"

]

for file in files:

    path = os.path.join(
        GRAPH_PATH,
        file
    )

    if os.path.exists(path):

        print(f"✓ {file}")

    else:

        print(f"✗ {file}")

  # ==========================================================
# FINAL SUMMARY
# ==========================================================

print("="*60)
print("NOTEBOOK 04 COMPLETED SUCCESSFULLY")
print("="*60)

print()

print("Test Accuracy : {:.2f}%".format(
    test_accuracy*100
))

print()

print("Evaluation Files Generated")

print("--------------------------")

print("✓ Confusion Matrix")

print("✓ Classification Report")

print("✓ ROC Curve")

print("✓ Accuracy Graph")

print("✓ Loss Graph")

print()

print("Graphs Saved In")

print(GRAPH_PATH)

print()

print("Next Notebook")

print("--------------------------")

print("Notebook 05 - Transfer Learning (MobileNetV2)")

print()

print("Project Progress")

print("--------------------------")

print("✓ Notebook 01 - Dataset Exploration")

print("✓ Notebook 02 - Data Preprocessing")

print("✓ Notebook 03 - Custom CNN Training")

print("✓ Notebook 04 - Model Evaluation")

print()







```
