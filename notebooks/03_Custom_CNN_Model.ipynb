```python
# ==========================================================
# NOTEBOOK 03
# CUSTOM CNN MODEL TRAINING
# ==========================================================

"""
OBJECTIVE
---------
This notebook builds and trains a Custom Convolutional Neural Network (CNN)
to classify Chest X-ray images into four classes.

Classes:
--------
1. COVID
2. Normal
3. Lung Opacity
4. Viral Pneumonia

The notebook performs the following tasks:

1. Load preprocessed datasets
2. Create TensorFlow data pipeline
3. Build a Custom CNN
4. Train the model
5. Save the best model
6. Save training history

Outputs
-------
best_custom_cnn.keras
history_custom.pkl

These files will be used later for model evaluation.
"""

# ==========================================================
# IMPORT LIBRARIES
# ==========================================================

import os
import shutil
import zipfile
import pickle

import numpy as np
import pandas as pd
import tensorflow as tf

from google.colab import drive

from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import (
    Conv2D,
    MaxPooling2D,
    BatchNormalization,
    GlobalAveragePooling2D,
    Dense,
    Dropout
)

from tensorflow.keras.callbacks import (
    EarlyStopping,
    ModelCheckpoint,
    ReduceLROnPlateau
)

print("="*60)
print("TensorFlow Version :", tf.__version__)
print("GPU :", tf.config.list_physical_devices("GPU"))
print("="*60)

# ==========================================================
# PROJECT SETTINGS
# ==========================================================

SEED = 42

tf.random.set_seed(SEED)

IMAGE_SIZE = (224,224)

BATCH_SIZE = 32

EPOCHS = 10

AUTOTUNE = tf.data.AUTOTUNE

# ==========================================================
# MOUNT GOOGLE DRIVE
# ==========================================================

drive.mount('/content/drive')

# ==========================================================
# PROJECT PATHS
# ==========================================================

PROJECT_PATH = "/content/drive/MyDrive/Covid_Major_Project"

ZIP_FILE = os.path.join(
    PROJECT_PATH,
    "dataset",
    "archive.zip"
)

PROCESSED_PATH = os.path.join(
    PROJECT_PATH,
    "processed"
)

MODEL_PATH = os.path.join(
    PROJECT_PATH,
    "models"
)

os.makedirs(MODEL_PATH, exist_ok=True)

TEMP_FOLDER="/content/COVID19_Dataset"

# ==========================================================
# EXTRACT DATASET
# ==========================================================

if os.path.exists(TEMP_FOLDER):
    shutil.rmtree(TEMP_FOLDER)

os.makedirs(TEMP_FOLDER,exist_ok=True)

print("\nExtracting Dataset...")

with zipfile.ZipFile(ZIP_FILE,"r") as zip_ref:
    zip_ref.extractall(TEMP_FOLDER)

print("Dataset Extracted Successfully.")

DATASET_PATH=os.path.join(
    TEMP_FOLDER,
    "COVID-19_Radiography_Dataset"
)

# ==========================================================
# LOAD PREPROCESSED FILES
# ==========================================================

train_df = pd.read_csv(
    os.path.join(PROCESSED_PATH,"train.csv")
)

validation_df = pd.read_csv(
    os.path.join(PROCESSED_PATH,"validation.csv")
)

test_df = pd.read_csv(
    os.path.join(PROCESSED_PATH,"test.csv")
)

with open(
    os.path.join(PROCESSED_PATH,"label_encoder.pkl"),
    "rb"
) as f:

    label_encoder = pickle.load(f)

class_weights = np.load(
    os.path.join(PROCESSED_PATH,"class_weights.npy"),
    allow_pickle=True
).item()

print("\nTraining Images :",len(train_df))
print("Validation Images :",len(validation_df))
print("Testing Images :",len(test_df))

print("\nLabel Mapping")

for cls in label_encoder.classes_:
    print(cls)

print("\nClass Weights")

print(class_weights)

# ==========================================================
# CREATE TENSORFLOW DATA PIPELINE
# ==========================================================

"""
This section creates TensorFlow datasets for training,
validation and testing.

The steps performed are:

1. Read image from disk
2. Decode PNG image
3. Resize image to 224 × 224
4. Normalize pixel values (0-255 → 0-1)
5. Batch the images
6. Prefetch batches for faster training
"""

# ==========================================================
# IMAGE PREPROCESSING FUNCTION
# ==========================================================

def preprocess_image(image_path, label):

    # Read image
    image = tf.io.read_file(image_path)

    # Decode PNG image into RGB format
    image = tf.image.decode_png(image, channels=3)

    # Resize image
    image = tf.image.resize(image, IMAGE_SIZE)

    # Normalize pixels
    image = tf.cast(image, tf.float32) / 255.0

    return image, label


# ==========================================================
# CREATE DATASET FUNCTION
# ==========================================================

def create_dataset(dataframe, training=False):

    dataset = tf.data.Dataset.from_tensor_slices(
        (
            dataframe["Image"].values,
            dataframe["Label_ID"].values
        )
    )

    dataset = dataset.map(
        preprocess_image,
        num_parallel_calls=AUTOTUNE
    )

    # Shuffle only the training dataset
    if training:
        dataset = dataset.shuffle(
            buffer_size=1000,
            seed=SEED
        )

    dataset = dataset.batch(BATCH_SIZE)

    dataset = dataset.prefetch(AUTOTUNE)

    return dataset


# ==========================================================
# CREATE TRAINING DATASET
# ==========================================================

train_dataset = create_dataset(
    train_df,
    training=True
)

# ==========================================================
# CREATE VALIDATION DATASET
# ==========================================================

validation_dataset = create_dataset(
    validation_df
)

# ==========================================================
# CREATE TEST DATASET
# ==========================================================

test_dataset = create_dataset(
    test_df
)

print("="*60)
print("TensorFlow Datasets Created Successfully")
print("="*60)

# ==========================================================
# VERIFY DATA PIPELINE
# ==========================================================

"""
Verify that images are loaded correctly.
"""

for images, labels in train_dataset.take(1):

    print("Image Batch Shape :", images.shape)

    print("Label Batch Shape :", labels.shape)

    print("Minimum Pixel Value :", tf.reduce_min(images).numpy())

    print("Maximum Pixel Value :", tf.reduce_max(images).numpy())

    print()

    print("Sample Labels")

    print(labels.numpy()[:10])

# ==========================================================
# DISPLAY DATASET INFORMATION
# ==========================================================

print("="*60)
print("DATASET INFORMATION")
print("="*60)

print("Training Samples   :", len(train_df))

print("Validation Samples :", len(validation_df))

print("Testing Samples    :", len(test_df))

print()

print("Image Size :", IMAGE_SIZE)

print("Batch Size :", BATCH_SIZE)

print("Classes :", len(label_encoder.classes_))

print()

print("Class Names")

for i, cls in enumerate(label_encoder.classes_):
    print(f"{i} -> {cls}")

# ==========================================================
# BUILD CUSTOM CNN MODEL
# ==========================================================

"""
CUSTOM CNN ARCHITECTURE

Input Image
     │
Conv2D (32 Filters)
     │
Batch Normalization
     │
Max Pooling
     │
Conv2D (64 Filters)
     │
Batch Normalization
     │
Max Pooling
     │
Conv2D (128 Filters)
     │
Batch Normalization
     │
Max Pooling
     │
Conv2D (256 Filters)
     │
Batch Normalization
     │
Max Pooling
     │
Global Average Pooling
     │
Dense (128)
     │
Dropout
     │
Dense (64)
     │
Dropout
     │
Output Layer (4 Classes)
"""

model = Sequential()

# ==========================================================
# INPUT LAYER
# ==========================================================

model.add(

    Conv2D(

        filters=32,

        kernel_size=(3,3),

        activation="relu",

        padding="same",

        input_shape=(224,224,3)

    )

)

model.add(BatchNormalization())

model.add(MaxPooling2D(pool_size=(2,2)))

# ==========================================================
# CONVOLUTION BLOCK 2
# ==========================================================

model.add(

    Conv2D(

        64,

        (3,3),

        activation="relu",

        padding="same"

    )

)

model.add(BatchNormalization())

model.add(MaxPooling2D((2,2)))

# ==========================================================
# CONVOLUTION BLOCK 3
# ==========================================================

model.add(

    Conv2D(

        128,

        (3,3),

        activation="relu",

        padding="same"

    )

)

model.add(BatchNormalization())

model.add(MaxPooling2D((2,2)))

# ==========================================================
# CONVOLUTION BLOCK 4
# ==========================================================

model.add(

    Conv2D(

        256,

        (3,3),

        activation="relu",

        padding="same"

    )

)

model.add(BatchNormalization())

model.add(MaxPooling2D((2,2)))

# ==========================================================
# FEATURE EXTRACTION
# ==========================================================

"""
GlobalAveragePooling drastically reduces the number of
parameters compared to Flatten(), making training faster
and reducing overfitting.
"""

model.add(GlobalAveragePooling2D())

# ==========================================================
# FULLY CONNECTED LAYERS
# ==========================================================

model.add(

    Dense(

        128,

        activation="relu"

    )

)

model.add(

    Dropout(0.30)

)

model.add(

    Dense(

        64,

        activation="relu"

    )

)

model.add(

    Dropout(0.30)

)

# ==========================================================
# OUTPUT LAYER
# ==========================================================

model.add(

    Dense(

        4,

        activation="softmax"

    )

)

# ==========================================================
# DISPLAY MODEL SUMMARY
# ==========================================================

print("="*60)
print("CUSTOM CNN MODEL")
print("="*60)

model.summary()

# ==========================================================
# COMPILE MODEL
# ==========================================================

"""
Compile the CNN using:

Optimizer : Adam
Loss      : Sparse Categorical Crossentropy
Metric    : Accuracy
"""

model.compile(

    optimizer="adam",

    loss="sparse_categorical_crossentropy",

    metrics=["accuracy"]

)

print()

print("="*60)

print("MODEL COMPILED SUCCESSFULLY")

print("="*60)

# ==========================================================
# TRAINING CALLBACKS
# ==========================================================

"""
Callbacks help improve training.

1. EarlyStopping
   Stops training if validation accuracy stops improving.

2. ModelCheckpoint
   Saves the best model automatically.

3. ReduceLROnPlateau
   Reduces learning rate when validation accuracy plateaus.
"""

BEST_MODEL = os.path.join(
    MODEL_PATH,
    "best_custom_cnn.keras"
)

early_stop = EarlyStopping(

    monitor="val_accuracy",

    patience=3,

    restore_best_weights=True,

    verbose=1

)

checkpoint = ModelCheckpoint(

    filepath=BEST_MODEL,

    monitor="val_accuracy",

    save_best_only=True,

    verbose=1

)

reduce_lr = ReduceLROnPlateau(

    monitor="val_loss",

    factor=0.5,

    patience=2,

    min_lr=1e-6,

    verbose=1

)

print("="*60)
print("CALLBACKS CREATED SUCCESSFULLY")
print("="*60)

# ==========================================================
# TRAIN THE MODEL
# ==========================================================

"""
Training Configuration

Epochs      : 10
Batch Size  : 32
Optimizer   : Adam
Loss        : Sparse Categorical Crossentropy

Expected Training Time
----------------------
Google Colab T4 GPU : 10–15 minutes
CPU                 : 25–35 minutes
"""

history = model.fit(

    train_dataset,

    validation_data=validation_dataset,

    epochs=10,

    class_weight=class_weights,

    callbacks=[

        early_stop,

        checkpoint,

        reduce_lr

    ],

    verbose=1

)

# ==========================================================
# SAVE TRAINING HISTORY
# ==========================================================

HISTORY_FILE = os.path.join(

    MODEL_PATH,

    "history_custom.pkl"

)

with open(HISTORY_FILE, "wb") as file:

    pickle.dump(history.history, file)

print()

print("="*60)
print("TRAINING HISTORY SAVED")
print("="*60)

print(HISTORY_FILE)

# ==========================================================
# DISPLAY TRAINING SUMMARY
# ==========================================================

print("="*60)
print("TRAINING COMPLETED")
print("="*60)

print("Best Model Saved :")

print(BEST_MODEL)

print()

print("History Saved :")

print(HISTORY_FILE)

print()

print("Best Validation Accuracy :")

print(max(history.history["val_accuracy"]))

print()

print("Final Training Accuracy :")

print(history.history["accuracy"][-1])

# ==========================================================
# IMPORT MATPLOTLIB
# ==========================================================

import matplotlib.pyplot as plt

print("="*60)
print("PLOTTING TRAINING CURVES")
print("="*60)

# ==========================================================
# ACCURACY CURVE
# ==========================================================

plt.figure(figsize=(8,5))

plt.plot(
    history.history["accuracy"],
    label="Training Accuracy",
    linewidth=2
)

plt.plot(
    history.history["val_accuracy"],
    label="Validation Accuracy",
    linewidth=2
)

plt.title("Training vs Validation Accuracy")

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.legend()

plt.grid(True)

plt.show()

# ==========================================================
# LOSS CURVE
# ==========================================================

plt.figure(figsize=(8,5))

plt.plot(
    history.history["loss"],
    label="Training Loss",
    linewidth=2
)

plt.plot(
    history.history["val_loss"],
    label="Validation Loss",
    linewidth=2
)

plt.title("Training vs Validation Loss")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.legend()

plt.grid(True)

plt.show()

# ==========================================================
# DISPLAY BEST RESULTS
# ==========================================================

best_epoch = np.argmax(history.history["val_accuracy"]) + 1

best_val_accuracy = max(history.history["val_accuracy"])

best_train_accuracy = history.history["accuracy"][best_epoch-1]

print("="*60)
print("BEST TRAINING RESULTS")
print("="*60)

print(f"Best Epoch                : {best_epoch}")

print(f"Training Accuracy         : {best_train_accuracy:.4f}")

print(f"Validation Accuracy       : {best_val_accuracy:.4f}")

print(f"Training Loss             : {history.history['loss'][best_epoch-1]:.4f}")

print(f"Validation Loss           : {history.history['val_loss'][best_epoch-1]:.4f}")

# ==========================================================
# SAVE TRAINING GRAPHS
# ==========================================================

GRAPH_FOLDER = os.path.join(
    PROJECT_PATH,
    "graphs"
)

os.makedirs(GRAPH_FOLDER, exist_ok=True)

# Accuracy Graph
plt.figure(figsize=(8,5))

plt.plot(history.history["accuracy"], linewidth=2)

plt.plot(history.history["val_accuracy"], linewidth=2)

plt.title("Training vs Validation Accuracy")

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.legend(["Training","Validation"])

plt.grid(True)

plt.savefig(
    os.path.join(
        GRAPH_FOLDER,
        "custom_accuracy.png"
    ),
    dpi=300,
    bbox_inches="tight"
)

plt.close()

# Loss Graph
plt.figure(figsize=(8,5))

plt.plot(history.history["loss"], linewidth=2)

plt.plot(history.history["val_loss"], linewidth=2)

plt.title("Training vs Validation Loss")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.legend(["Training","Validation"])

plt.grid(True)

plt.savefig(
    os.path.join(
        GRAPH_FOLDER,
        "custom_loss.png"
    ),
    dpi=300,
    bbox_inches="tight"
)

plt.close()

print()

print("Graphs saved successfully.")

print(GRAPH_FOLDER)

# ==========================================================
# VERIFY SAVED FILES
# ==========================================================

print("="*60)
print("VERIFYING SAVED FILES")
print("="*60)

files = [

    "best_custom_cnn.keras",

    "history_custom.pkl"

]

for file in files:

    path = os.path.join(MODEL_PATH, file)

    if os.path.exists(path):

        print(f"✓ {file}")

    else:

        print(f"✗ {file}")


# ==========================================================
# DISPLAY MODEL INFORMATION
# ==========================================================

print("="*60)
print("MODEL INFORMATION")
print("="*60)

print("Model Name        : Custom CNN")

print("Input Size        :", IMAGE_SIZE)

print("Batch Size        :", BATCH_SIZE)

print("Epochs            :", len(history.history["accuracy"]))

print("Optimizer         : Adam")

print("Loss Function     : Sparse Categorical Crossentropy")

print("Output Classes    :", len(label_encoder.classes_))

# ==========================================================
# DISPLAY FINAL PERFORMANCE
# ==========================================================

best_epoch = np.argmax(history.history["val_accuracy"]) + 1

print("="*60)
print("FINAL TRAINING RESULTS")
print("="*60)

print("Best Epoch                :", best_epoch)

print("Training Accuracy         : {:.4f}".format(

    history.history["accuracy"][best_epoch-1]

))

print("Validation Accuracy       : {:.4f}".format(

    history.history["val_accuracy"][best_epoch-1]

))

print("Training Loss             : {:.4f}".format(

    history.history["loss"][best_epoch-1]

))

print("Validation Loss           : {:.4f}".format(

    history.history["val_loss"][best_epoch-1]

))

# ==========================================================
# SAVE MODEL SUMMARY
# ==========================================================

summary_file = os.path.join(

    MODEL_PATH,

    "custom_cnn_summary.txt"

)

with open(summary_file, "w") as f:

    model.summary(

        print_fn=lambda x: f.write(x + "\n")

    )

print("Model Summary Saved Successfully.")

print(summary_file)

# ==========================================================
# NOTEBOOK SUMMARY
# ==========================================================

print("="*60)
print("NOTEBOOK 03 COMPLETED SUCCESSFULLY")
print("="*60)

print()

print("Dataset Used")

print("----------------------------")

print("Training Images   :", len(train_df))

print("Validation Images :", len(validation_df))

print("Testing Images    :", len(test_df))

print()

print("Model Files")

print("----------------------------")

print("✓ best_custom_cnn.keras")

print("✓ history_custom.pkl")

print("✓ custom_cnn_summary.txt")

print()

print("Graphs")

print("----------------------------")

print("✓ custom_accuracy.png")

print("✓ custom_loss.png")

print()

print("Next Notebook")

print("----------------------------")

print("Notebook 04 - Model Evaluation")

print()

print("Project Status :")

print("Notebook 01 ✓ Completed")

print("Notebook 02 ✓ Completed")

print("Notebook 03 ✓ Completed")


```
