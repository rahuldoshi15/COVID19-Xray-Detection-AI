```python
# ==========================================================
# NOTEBOOK 05
# TRANSFER LEARNING USING MOBILENETV2
# ==========================================================

"""
OBJECTIVE
---------
This notebook uses MobileNetV2 as a pre-trained model
for COVID-19 X-ray classification.

Steps:

1. Load processed dataset
2. Build MobileNetV2 model
3. Train only the classification head
4. Save best model
5. Plot training graphs
"""

# ==========================================================
# IMPORT LIBRARIES
# ==========================================================

import os
import pickle
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from google.colab import drive

from sklearn.utils.class_weight import compute_class_weight

from tensorflow.keras.applications import MobileNetV2

from tensorflow.keras.models import Model

from tensorflow.keras.layers import (

    Input,

    GlobalAveragePooling2D,

    Dense,

    Dropout

)

from tensorflow.keras.callbacks import (

    EarlyStopping,

    ReduceLROnPlateau,

    ModelCheckpoint

)

print("="*60)

print("TensorFlow :", tf.__version__)

print("GPU :", tf.config.list_physical_devices("GPU"))

print("="*60)

# ==========================================================
# SETTINGS
# ==========================================================

IMAGE_SIZE = (224,224)

BATCH_SIZE = 32

EPOCHS = 10

AUTOTUNE = tf.data.AUTOTUNE

SEED = 42

tf.random.set_seed(SEED)

np.random.seed(SEED)

# ==========================================================
# MOUNT GOOGLE DRIVE
# ==========================================================

drive.mount("/content/drive")

# ==========================================================
# PROJECT PATHS
# ==========================================================

PROJECT_PATH="/content/drive/MyDrive/Covid_Major_Project"

PROCESSED_PATH=os.path.join(

    PROJECT_PATH,

    "processed"

)

MODEL_PATH=os.path.join(

    PROJECT_PATH,

    "models"

)

GRAPH_PATH=os.path.join(

    PROJECT_PATH,

    "graphs"

)

os.makedirs(MODEL_PATH,exist_ok=True)

os.makedirs(GRAPH_PATH,exist_ok=True)

# ==========================================================
# LOAD DATASET
# ==========================================================

train_df=pd.read_csv(

    os.path.join(

        PROCESSED_PATH,

        "train.csv"

    )

)

validation_df=pd.read_csv(

    os.path.join(

        PROCESSED_PATH,

        "validation.csv"

    )

)

test_df=pd.read_csv(

    os.path.join(

        PROCESSED_PATH,

        "test.csv"

    )

)

print()

print("="*60)

print("DATASET LOADED SUCCESSFULLY")

print("="*60)

print("Training Images   :",len(train_df))

print("Validation Images :",len(validation_df))

print("Testing Images    :",len(test_df))

# ==========================================================
# LOAD CLASS WEIGHTS
# ==========================================================

class_weights=np.load(

    os.path.join(

        PROCESSED_PATH,

        "class_weights.npy"

    ),

    allow_pickle=True

).item()

print()

print("Class Weights")

print(class_weights)

# ==========================================================
# LOAD LABEL ENCODER
# ==========================================================

with open(

    os.path.join(

        PROCESSED_PATH,

        "label_encoder.pkl"

    ),

    "rb"

) as file:

    label_encoder=pickle.load(file)

print()

print("Classes")

print(label_encoder.classes_)

# ==========================================================
# EXTRACT DATASET
# ==========================================================

import shutil
import zipfile

ZIP_FILE = os.path.join(
    PROJECT_PATH,
    "dataset",
    "archive.zip"
)

TEMP_FOLDER = "/content/COVID19_Dataset"

if os.path.exists(TEMP_FOLDER):
    shutil.rmtree(TEMP_FOLDER)

os.makedirs(TEMP_FOLDER, exist_ok=True)

print("="*60)
print("EXTRACTING DATASET")
print("="*60)

with zipfile.ZipFile(ZIP_FILE, "r") as zip_ref:
    zip_ref.extractall(TEMP_FOLDER)

print("Dataset Extracted Successfully.")

DATASET_PATH = os.path.join(
    TEMP_FOLDER,
    "COVID-19_Radiography_Dataset"
)

print(DATASET_PATH)

# ==========================================================
# IMAGE PREPROCESSING FUNCTION
# ==========================================================

from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

def preprocess_image(image_path, label):

    image = tf.io.read_file(image_path)

    image = tf.image.decode_png(
        image,
        channels=3
    )

    image = tf.image.resize(
        image,
        IMAGE_SIZE
    )

    image = preprocess_input(image)

    return image, label

# ==========================================================
# CREATE TENSORFLOW DATASETS
# ==========================================================

train_dataset = tf.data.Dataset.from_tensor_slices(

    (
        train_df["Image"].values,
        train_df["Label_ID"].values
    )

)

validation_dataset = tf.data.Dataset.from_tensor_slices(

    (
        validation_df["Image"].values,
        validation_df["Label_ID"].values
    )

)

test_dataset = tf.data.Dataset.from_tensor_slices(

    (
        test_df["Image"].values,
        test_df["Label_ID"].values
    )

)

# ==========================================================
# APPLY PREPROCESSING
# ==========================================================

train_dataset = train_dataset.map(

    preprocess_image,

    num_parallel_calls=AUTOTUNE

)

validation_dataset = validation_dataset.map(

    preprocess_image,

    num_parallel_calls=AUTOTUNE

)

test_dataset = test_dataset.map(

    preprocess_image,

    num_parallel_calls=AUTOTUNE
)

# ==========================================================
# BATCH, SHUFFLE AND PREFETCH
# ==========================================================

train_dataset = train_dataset.shuffle(

    1000,

    seed=SEED

)

train_dataset = train_dataset.batch(BATCH_SIZE)

validation_dataset = validation_dataset.batch(BATCH_SIZE)

test_dataset = test_dataset.batch(BATCH_SIZE)

train_dataset = train_dataset.prefetch(AUTOTUNE)

validation_dataset = validation_dataset.prefetch(AUTOTUNE)

test_dataset = test_dataset.prefetch(AUTOTUNE)

print()

print("="*60)
print("DATASETS CREATED SUCCESSFULLY")
print("="*60)

print("Training Images   :", len(train_df))

print("Validation Images :", len(validation_df))

print("Testing Images    :", len(test_df))

print("\nFirst training image path:")
print(train_df["Image"].iloc[0])

import os
print("\nDoes this file exist?")
print(os.path.exists(train_df["Image"].iloc[0]))

# ==========================================================
# VERIFY DATASET
# ==========================================================

print("="*60)
print("VERIFY DATASET")
print("="*60)

for images, labels in train_dataset.take(1):

    print("Image Batch Shape :", images.shape)

    print("Label Batch Shape :", labels.shape)

    print("Minimum Pixel Value :", tf.reduce_min(images).numpy())

    print("Maximum Pixel Value :", tf.reduce_max(images).numpy())

# ==========================================================
# VERIFY DATASET
# ==========================================================

print("="*60)
print("VERIFY DATASET")
print("="*60)

for images, labels in train_dataset.take(1):

    print("Image Batch Shape :", images.shape)

    print("Label Batch Shape :", labels.shape)

    print("Minimum Pixel Value :", tf.reduce_min(images).numpy())

    print("Maximum Pixel Value :", tf.reduce_max(images).numpy())

# ==========================================================
# LOAD PRE-TRAINED MOBILENETV2
# ==========================================================

"""
Load MobileNetV2 with ImageNet weights.

The convolutional base is frozen so that only the
new classification layers are trained.
"""

base_model = MobileNetV2(

    weights="imagenet",

    include_top=False,

    input_shape=(224,224,3)

)

base_model.trainable = False

print("="*60)
print("MOBILENETV2 LOADED")
print("="*60)

print("Trainable :", base_model.trainable)

# ==========================================================
# BUILD TRANSFER LEARNING MODEL
# ==========================================================

inputs = Input(shape=(224,224,3))

x = base_model(

    inputs,

    training=False

)

x = GlobalAveragePooling2D()(x)

x = Dense(

    256,

    activation="relu"

)(x)

x = Dropout(0.30)(x)

x = Dense(

    128,

    activation="relu"

)(x)

x = Dropout(0.20)(x)

outputs = Dense(

    4,

    activation="softmax"

)(x)

model = Model(

    inputs,

    outputs

)

# ==========================================================
# COMPILE MODEL
# ==========================================================

model.compile(

    optimizer=tf.keras.optimizers.Adam(

        learning_rate=0.0001

    ),

    loss="sparse_categorical_crossentropy",

    metrics=["accuracy"]

)

print("="*60)
print("MODEL COMPILED SUCCESSFULLY")
print("="*60)

# ==========================================================
# MODEL SUMMARY
# ==========================================================

model.summary()

# ==========================================================
# TRAINABLE PARAMETERS
# ==========================================================

trainable = np.sum(

    [np.prod(v.shape) for v in model.trainable_weights]

)

non_trainable = np.sum(

    [np.prod(v.shape) for v in model.non_trainable_weights]

)

print()

print("="*60)
print("MODEL INFORMATION")
print("="*60)

print("Trainable Parameters     :", trainable)

print("Non-Trainable Parameters :", non_trainable)

print("Total Parameters         :", trainable + non_trainable)

# ==========================================================
# CREATE CALLBACKS
# ==========================================================

BEST_MODEL = os.path.join(

    MODEL_PATH,

    "best_mobilenetv2.keras"

)

early_stop = EarlyStopping(

    monitor="val_accuracy",

    patience=3,

    restore_best_weights=True,

    verbose=1

)

checkpoint = ModelCheckpoint(

    filepath=BEST_MODEL,

    monitor="val_accuracy",

    save_best_only=True,

    verbose=1

)

reduce_lr = ReduceLROnPlateau(

    monitor="val_loss",

    factor=0.5,

    patience=2,

    min_lr=1e-6,

    verbose=1

)

print("="*60)

print("CALLBACKS CREATED SUCCESSFULLY")

print("="*60)

# ==========================================================
# TRAIN THE MODEL
# ==========================================================

print("="*60)

print("STARTING TRAINING")

print("="*60)

history = model.fit(

    train_dataset,

    validation_data=validation_dataset,

    epochs=10,

    class_weight=class_weights,

    callbacks=[

        early_stop,

        checkpoint,

        reduce_lr

    ],

    verbose=1

)

# ==========================================================
# SAVE TRAINING HISTORY
# ==========================================================

history_file = os.path.join(

    MODEL_PATH,

    "history_mobilenetv2.pkl"

)

with open(history_file,"wb") as file:

    pickle.dump(

        history.history,

        file

    )

print()

print("Training History Saved Successfully.")

print(history_file)

# ==========================================================
# DISPLAY BEST RESULTS
# ==========================================================

best_epoch = np.argmax(

    history.history["val_accuracy"]

) + 1

print("="*60)

print("TRAINING RESULTS")

print("="*60)

print("Best Epoch :", best_epoch)

print()

print("Training Accuracy : {:.4f}".format(

    history.history["accuracy"][best_epoch-1]

))

print("Validation Accuracy : {:.4f}".format(

    history.history["val_accuracy"][best_epoch-1]

))

print("Training Loss : {:.4f}".format(

    history.history["loss"][best_epoch-1]

))

print("Validation Loss : {:.4f}".format(

    history.history["val_loss"][best_epoch-1]

))

# ==========================================================
# VERIFY SAVED MODEL
# ==========================================================

print()

print("="*60)

print("VERIFY MODEL")

print("="*60)

if os.path.exists(BEST_MODEL):

    print("✓ Best MobileNetV2 Model Saved")

else:

    print("✗ Model Not Found")

if os.path.exists(history_file):

    print("✓ Training History Saved")

else:

    print("✗ History Not Found")

# ==========================================================
# IMPORT MATPLOTLIB
# ==========================================================

import matplotlib.pyplot as plt

print("="*60)
print("PLOTTING TRAINING CURVES")
print("="*60)

# ==========================================================
# TRAINING ACCURACY GRAPH
# ==========================================================

plt.figure(figsize=(8,5))

plt.plot(

    history.history["accuracy"],

    linewidth=2,

    label="Training Accuracy"

)

plt.plot(

    history.history["val_accuracy"],

    linewidth=2,

    label="Validation Accuracy"

)

plt.title("MobileNetV2 Training Accuracy")

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.legend()

plt.grid(True)

plt.tight_layout()

plt.show()

# ==========================================================
# TRAINING LOSS GRAPH
# ==========================================================

plt.figure(figsize=(8,5))

plt.plot(

    history.history["loss"],

    linewidth=2,

    label="Training Loss"

)

plt.plot(

    history.history["val_loss"],

    linewidth=2,

    label="Validation Loss"

)

plt.title("MobileNetV2 Training Loss")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.legend()

plt.grid(True)

plt.tight_layout()

plt.show()

# ==========================================================
# SAVE ACCURACY GRAPH
# ==========================================================

plt.figure(figsize=(8,5))

plt.plot(

    history.history["accuracy"],

    linewidth=2

)

plt.plot(

    history.history["val_accuracy"],

    linewidth=2

)

plt.title("MobileNetV2 Accuracy")

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.legend([

    "Training",

    "Validation"

])

plt.grid(True)

plt.tight_layout()

plt.savefig(

    os.path.join(

        GRAPH_PATH,

        "mobilenet_accuracy.png"

    ),

    dpi=300,

    bbox_inches="tight"

)

plt.close()

print("Accuracy Graph Saved Successfully.")

# ==========================================================
# SAVE LOSS GRAPH
# ==========================================================

plt.figure(figsize=(8,5))

plt.plot(

    history.history["loss"],

    linewidth=2

)

plt.plot(

    history.history["val_loss"],

    linewidth=2

)

plt.title("MobileNetV2 Loss")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.legend([

    "Training",

    "Validation"

])

plt.grid(True)

plt.tight_layout()

plt.savefig(

    os.path.join(

        GRAPH_PATH,

        "mobilenet_loss.png"

    ),

    dpi=300,

    bbox_inches="tight"

)

plt.close()

print("Loss Graph Saved Successfully.")

# ==========================================================
# DISPLAY BEST RESULTS
# ==========================================================

best_epoch = np.argmax(

    history.history["val_accuracy"]

) + 1

print("="*60)

print("BEST TRAINING RESULTS")

print("="*60)

print(f"Best Epoch                : {best_epoch}")

print(f"Training Accuracy         : {history.history['accuracy'][best_epoch-1]:.4f}")

print(f"Validation Accuracy       : {history.history['val_accuracy'][best_epoch-1]:.4f}")

print(f"Training Loss             : {history.history['loss'][best_epoch-1]:.4f}")

print(f"Validation Loss           : {history.history['val_loss'][best_epoch-1]:.4f}")

print()

print("Graphs saved successfully.")

# ==========================================================
# VERIFY SAVED FILES
# ==========================================================

print("="*60)
print("VERIFYING SAVED FILES")
print("="*60)

files = [

    "best_mobilenetv2.keras",

    "history_mobilenetv2.pkl"

]

for file in files:

    path = os.path.join(MODEL_PATH, file)

    if os.path.exists(path):

        print(f"✓ {file}")

    else:

        print(f"✗ {file}")

# ==========================================================
# SAVE MODEL SUMMARY
# ==========================================================

summary_file = os.path.join(

    MODEL_PATH,

    "mobilenetv2_summary.txt"

)

with open(summary_file, "w") as f:

    model.summary(

        print_fn=lambda x: f.write(x + "\n")

    )

print()

print("Model Summary Saved Successfully.")

print(summary_file)

# ==========================================================
# DISPLAY MODEL INFORMATION
# ==========================================================

print("="*60)
print("MODEL INFORMATION")
print("="*60)

print("Model Name        : MobileNetV2")

print("Input Size        :", IMAGE_SIZE)

print("Batch Size        :", BATCH_SIZE)

print("Epochs Trained    :", len(history.history["accuracy"]))

print("Optimizer         : Adam")

print("Learning Rate     : 0.0001")

print("Output Classes    :", len(label_encoder.classes_))

print()

print("Classes")

for cls in label_encoder.classes_:

    print("•", cls)

# ==========================================================
# DISPLAY FINAL RESULTS
# ==========================================================

best_epoch = np.argmax(

    history.history["val_accuracy"]

) + 1

print("="*60)
print("FINAL TRAINING RESULTS")
print("="*60)

print("Best Epoch :", best_epoch)

print()

print("Training Accuracy : {:.2f}%".format(

    history.history["accuracy"][best_epoch-1]*100

))

print("Validation Accuracy : {:.2f}%".format(

    history.history["val_accuracy"][best_epoch-1]*100

))

print("Training Loss : {:.4f}".format(

    history.history["loss"][best_epoch-1]

))

print("Validation Loss : {:.4f}".format(

    history.history["val_loss"][best_epoch-1]

))

# ==========================================================
# NOTEBOOK SUMMARY
# ==========================================================

print("="*60)
print("NOTEBOOK 05 COMPLETED SUCCESSFULLY")
print("="*60)

print()

print("Dataset")

print("----------------------------")

print("Training Images   :", len(train_df))

print("Validation Images :", len(validation_df))

print("Testing Images    :", len(test_df))

print()

print("Files Generated")

print("----------------------------")

print("✓ best_mobilenetv2.keras")

print("✓ history_mobilenetv2.pkl")

print("✓ mobilenetv2_summary.txt")

print("✓ mobilenet_accuracy.png")

print("✓ mobilenet_loss.png")

print()

print("Saved In")

print("----------------------------")

print(MODEL_PATH)

print()

print("Next Notebook")

print("----------------------------")

print("Notebook 06 - Model Comparison")

print()

print("Project Progress")

print("----------------------------")


```
